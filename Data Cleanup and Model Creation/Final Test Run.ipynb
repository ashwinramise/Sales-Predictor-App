```python
def ExtraTreesR(Xtrain, Xtest, Ytrain, Ytest):
    model = ExtraTreesRegressor(n_estimators=10,random_state=10, bootstrap=True)
    model.fit(Xtrain,Ytrain)
    yPred = model.predict(Xtest)
    ytest = Ytest.to_list()
    diff = [abs(ytest[i] - yPred[i]) for i in range(len(ytest))]
    d0 = [ele for ele in diff if ele==0]
    d10 = [ele for ele in diff if ele<10]
    d20 = [ele for ele in diff if ele<20]
    Countac = sum(yPred)/sum(ytest)
    ETRac0 = len(d0)/len(diff)
    ETRac10 = len(d10)/len(diff)
    ETRac20 = len(d20)/len(diff)
    ETRac = [ETRac0,ETRac10,ETRac20, Countac]
    ETR = model.feature_importances_
    Feature_importances['ETR'] = ETR
    Acc['ETR'] = ETRac
```

```python
def writecols(df,state,county,brand=None,subbrand=None):
    if brand == None and subbrand==None:
        df['Bill Date']=pd.to_datetime(df['Bill Date'],dayfirst=True)
        df['State'] = State.get(state)
        df['County'] = County.get(county)
        df['Months'] = [date.month for date in df['Bill Date'].to_list()]
        df['Year'] = [date.year for date in df['Bill Date'].to_list()]
    else:
        df['Bill Date']=pd.to_datetime(df['Bill Date'],dayfirst=True)
        df['State'] = State.get(state)
        df['County'] = County.get(county)
        df['Months'] = [date.month for date in df['Bill Date'].to_list()]
        df['Year'] = [date.year for date in df['Bill Date'].to_list()]
        df['Brand'] = Brand.get(brand)
        df['Sub Brand']= SubBrand.get(subbrand)
    return df

def predictor(df,state,county,brand=None,subbrand=None,method):
    '''Mehtod = Natural_Hist/PandemicRegressor'''
    df = writecols(df,state,county,brand,subbrand)
    struct_data = df.copy()
    non_numeric_columns = list(struct_data.select_dtypes(exclude=[_np.number]).columns)
    le = LabelEncoder()
    for col in non_numeric_columns:
        if col in struct_data.columns:
            i = struct_data.columns.get_loc(col)
            struct_data.iloc[:, i] = struct_data.apply(lambda i: le.fit_transform(i.astype(str)), axis=0,
                                                       result_type='expand')
    if method == 'Natural_Hist':
        DataFrame = struct_data[['State', 'County', 'Bill Date', 'Months']]
        model= joblib.load(f'.\DataModels\State\Predict Sales\{state}\History_Sales.joblib')
        Preds = model.predict(Dataframe)
        df['Nhist'] = Preds
        return df
    elif method == 'PandemicRegressor':
        DataFrame = struct_data[
                ['State', 'County', 'Brand', 'Sub Brand', 'Bill Date',
       'Covid cases', 'Months', 'Year']]
        model=joblib.load('.\DataModels\State\Predict Sales\ETR.joblib')
        df['PandemicSales'] = model.predict(DataFrame)
        return df
```
